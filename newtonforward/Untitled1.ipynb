```python
import numpy as np
import pandas as pd
import datetime
```

```python
def u_call(u,n):
    temp=u
    for i in range(1,n):
        temp=temp*(u-i)
    return temp
def fact(n):
    f=1
    for i in range(2,n+1):
        f=f*i
    return f
n=9   
```

```python
for l in range(0,4500):
    df=pd.read_excel(r'data3.xlsx')
    a=np.array(df)
    xlist=list(range(0,160,20))
    xlist
    ylist=a[l,0::2]
    x=np.array(xlist)
    y=np.array(ylist)
    y_mat=np.zeros((n,n),dtype=float)

    for i in range(n):
        y_mat[i][0]=float(y[i])

    for i in range(1,n):
        for j in range(n-i):
            y_mat[j][i]=y_mat[j+1][i-1]-y_mat[j][i-1]


    #lets consider that we want to know the value of x for y=32
    b=[]
    value=[0,10,30,50,70,90,100,110,130,150]
    for i in value:
        sum=y_mat[0][0]
        u=(value-x[0])/(x[1]-x[0])
        for i in range(1,n):
            sum=sum+(u_call(u,i)*y_mat[0][i])/fact(i)
    b.append(sum)
    df = pd.DataFrame(b)
    mod=pd.read_csv(r'data3.csv')
    with open('data3.csv', 'a') as f:
             (df).to_csv(f, header=False)
    print(datetime.datetime.time(datetime.datetime.now()),end='')
```
